```
import numpy as np
import matplotlib.pyplot as plt
```

```
def relu(x):
  return (np.maximum(x,0))

# x = np.linspace(-9.9,10,200)
x = np.linspace(-1.9,2,200)
```

```
plt.plot(x,relu(x))
plt.show()
```

approximate function: f1 = x**3+x**2+x+2
with single hidden layer having relu activation fn.

```
f1 = x**3+x**2+x+2
plt.plot(x,f1)
plt.show()
```

```
n1 = relu(-6*x-7.7)
n2 = relu(-1.3*x-1.3)
n3 = relu(2*x+1)
n4 = relu(1.2*x+1.6)
n5 = relu(2*x+1)
n6 = relu(6.5*x-5)
plt.plot(x,n1,color='blue')
plt.plot(x,n2,color='green')
plt.plot(x,n3,color='orange')
plt.plot(x,n4,color='yellow')
plt.plot(x,n5,color='red')
plt.plot(x,n6,color='black')
plt.show()

```

```
z= -n1-n2-n3+n4+n5+n6
plt.plot(x,z)
plt.plot(x,f1)
plt.show()
```
