# Transfer Functions

## Linear

```

def linear(x):
  return x 
```

```
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, linear(x))
```

## ReLu

```
# ReLu

def reLu(x):
  return np.array([xi if xi >= 0 else 0 for xi in x])
```

```
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, reLu(x))
```

# LeakyReLu

```
def leakyReLu(x, negative_slope=0.01):
  return np.array([xi if xi>=0 else xi*negative_slope for xi in x])
```

```
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, leakyReLu(x))
```

```
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, leakyReLu(x, negative_slope=0.1))
```

## Sigmoid

```
def sigmoid(x, lam=1):
  return 1/(1+np.exp(-lam*x))
```

```
x=np.arange(-10,10,0.005)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, sigmoid(x))
```

```
# With a different lambda
x=np.arange(-10,10,0.005)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, sigmoid(x,10))
```

## Exponential Linear Unit (ELU)

```
def eLU(x, alpha=1):
  return np.array([xi if xi>0 else alpha*(np.exp(xi)-1) for xi in x ])
```

```
x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, eLU(x))
```

```
# With a different alpha
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, eLU(x,5))
```

## LogSigmoid

```
def logsigmoid(x, lam=1):
  return np.log(1/(1+np.exp(-lam*x)))
```

```

x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, logsigmoid(x))
```

```
# With a different lambda
x=np.arange(-10,10,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, logsigmoid(x,5))
```

## Tanh

```
def tanh(x, lam=1):
  return (np.exp(lam*x)-np.exp(-lam*x))/(np.exp(lam*x)+np.exp(-lam*x))
```

```
# With a different lambda
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, tanh(x))
```

```
# With a different lambda
x=np.arange(-2,2,0.01)
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x, tanh(x, 5))
```

# Swish (or SiLu, Sigmoid Linear unit)
- It was introduced by Google.
- It performs better than ReLU.
- It is non-monotonic.

```
def silu(x,beta=1):
  return x*sigmoid(beta*x)
```

```
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x,silu(x))
```

```
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x,silu(x, beta=100))
```

# Softmax
- For softmax, value at a point is dependent on all other points
- Sum of all values is 1
- Converts to probabilities

```
def softmax(x):
  return np.exp(x)/np.exp(x).sum()
```

```
plt.axhline(0,color='grey') 
plt.axvline(0,color='grey') 
plt.plot(x,softmax(x))
```
