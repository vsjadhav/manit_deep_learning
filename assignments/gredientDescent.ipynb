```
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```
x = np.linspace(-180,180,1000)
f0 = (4*np.sin(x)*x**9)-(9*x**7)-(36*x**6)
f1 = (29*x**2)-(97*x)+77
f2 = (x**2)+ x**2*(2+np.sin(x/5)) 
# plt.plot(x,f0)
# plt.plot(x,f1)
plt.plot(x,f2)
plt.show()
```

```
f = lambda x: (x**2)+ x**2*(2+np.sin(x/5)) 
df = lambda x: 6*x + (((2*x*np.sin(x/5))+((x**2)*(np.cos(x/5))/5))/((np.sin(x/5))**2))
previous_step_size = 1
prev_x = None
cur_x = np.random.randint(360) - 180
print(f"variable initialised with: {cur_x}")
iters = 0
max_iters = 20000
while previous_step_size != 0 and iters < max_iters:
  prev_x = cur_x
  cur_x = cur_x - 0.0001*df(cur_x)          #Gradient Descent
  previous_step_size = abs(cur_x - prev_x)
  iters = iters + 1
  
plt.plot(x,f(x),'b')
plt.plot(cur_x,f(cur_x),'r+')
plt.show()
print(iters)
print("The local minima occures at",cur_x)
```
